```python
#Create separate .csv files per genre
#
import json
import pandas as pd
def clean_lyrics_helper(lyric_string):
    lyric_string=lyric_string.strip()
    messy_characters=[",","'","[",']',"(",")"]
    #Start checking from beginning to end
    flag1=1
    #print("2-",lyric_string[1:])
    while(flag1):
        if lyric_string!="":
            if lyric_string[0] in messy_characters:
                lyric_string=lyric_string[1:]
            else:
                flag1=0
        else:
            flag1=0
    #Start checking from end to beginning
    flag2=1
    while(flag2):
        if(lyric_string!=""):
            if lyric_string[-1] in messy_characters:
                lyric_string=lyric_string[:-1]
            else:
                flag2=0
        else:
            flag2=0
    return lyric_string

def clean_lyrics(lyric_string):
    lyric_string=lyric_string.strip()
    l=lyric_string.split(" ")
    tmp=""
    for i in range(0,len(l)):
        #print("1-",lyric_string,l[i])
        cleaned=clean_lyrics_helper(l[i])
        #print(cleaned)
        tmp=tmp+ cleaned
        tmp=tmp+" "
    lyric_string=tmp.strip()
    return lyric_string
    
#Load the datasets
lyric_summaries = pd.read_pickle(r'id_to_summary_lines.pickle')
df_lyrics= pd.DataFrame(lyric_summaries.items(), columns=['song_id', 'lyric_summaries'])
df_main = pd.read_csv('songs_genres_grouped_2.csv')

list_genres=["pop","reggae","metal","hip-hop","blues","dance_music",
            "electronic","poetry","jazz","rock","country","funk",
             "house","religious","celtic","theatre",
            "rap","folk","soul","punk","new_wave","psychedelic",
            "christmas","trance","techno","latin","contemporary",
            "grunge","classical_music","hardcore","harmony" ,
            "acoustic","comedy_songs","instrumental"]

for genre in list_genres:
    df_temp = df_main[df_main["grouped_genre"]==genre]
    df_temp_lyrics = pd.merge(df_temp, df_lyrics, how='inner', on=['song_id'])
    df_temp_lyrics.drop_duplicates(subset ="song_id",
                     keep ="first", inplace = True,ignore_index=True)
    dff = pd.DataFrame()
    dff = pd.DataFrame(columns=['year', 'V1'])

    for i in range(0,len(df_temp_lyrics)):
        all_lyrics=""
        for j in range(0,len(df_temp_lyrics["lyric_summaries"][i])):
            all_lyrics=all_lyrics+clean_lyrics(df_temp_lyrics["lyric_summaries"][i][j])
            all_lyrics=all_lyrics+" "

        dff = dff.append({'year':df_temp_lyrics["publicationDate"][i] ,'V1': all_lyrics}, ignore_index=True)

    print("finished ",genre)
    import pandas as pd
    dff.to_csv('GENRES/'+str(genre)+'.csv', index=False)



```
